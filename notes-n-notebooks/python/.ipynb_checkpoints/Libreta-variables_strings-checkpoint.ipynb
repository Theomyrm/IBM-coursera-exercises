```python
# Most used variable types on python are:

int(1) #Integer; used for store whole values, with no decimals.
float(1.0) #Float; store numbers with decimals.
str(12) #String; stores literal characters, taking no numeric value, used to store strings of characters.
bool(1) #Bool; stores logic values True, False.

# Useful methods

# 1. type(x) --> used for printing the type of variable x
## Examples:

x = 15.0 # The .0 still counts as float. So ->
# The type of x is float
type(x)

y = 13 # This number has no decimals ->
# The type of y is int
type(y)

z = "This is a string" # Stores words and characters
# The type of z is string
type(z)

w = False # Store boolean value (True or False)
# The type of w is bool
type(w)

# 2. var(x) --> convert a type of var to other.
## Examples:

str(x) # converts 15.0 to "15.0" string

int(x) # converts 15.0 to 15 int
```

```python
"""

#  #  #  #  #  #

#  Operations: #

#  Arithmetics #

#  String Ops  #

#  #  #  #  #  #


"""

# Arithmetic operations:

2  + 3  # +  = adition
2  - 3  # -  = substraction
2  * 3  # *  = multiplication
2  / 3  # /  = division
2 // 3  # // = floor division
2 ** 3  # ** = exponentiation
2  % 3  # %  = modulus



# Declare a string

str1 = "abcdefg" ## Each character has an index

#   [0,1,2,3,4,5,6] each number represents a place on the string
                    # This can be aplied backwise
#   [-7,-6,-5,-4,-3,-2,-1] this is counted from end to beginning 
        
       # Examples:
        
c1 = str1[0]  # this is = a
print(c1)
c2 = str1[1]  # this is = b
print(c2)
c3 = str1[2]  # this is = c
print(c3)
c7 = str1[-1] # this is = g
print(c7)
c6 = str1[-2] # this is = f
print(c6)
c5 = str1[-3] # this is = e
print(c5)

# There is also the possibility of using a range of elements of a string
# This can be done using the ':' simbol to specify a range. For example:

       # Examples:

c1_3 = str1[0:3]   # this is = abc
print(c1_3)

# Other things, like using a stride may have different results:

c1_3_5_7 = str1[::2]  # this is = aceg
print(c1_3_5_7)
c2_4_6 = str1[1:7:2] # this is = bdf
#            [1(starting point):7(lenght of interval):2(steps)]
print(c2_4_6)

# Function len() returns number of characters on string:

print("The total number of characters on str1 string are", len(str1))

# There are many ways of manipulating strings:

# Adition of strings with other strings

statement = str1 + " Are the first 7 letters of the alphabet"
print(statement)

# Strings can be also multiplied by this logic:

str2 = str1*2 # This is = abcdefgabcdefg
print(str2)
```

```python
"""

#  #  #  #  #  #

#  Strings:    #

#  Format      #

#  Readability #

#  Methods     #

#  #  #  #  #  #

"""

# There are signs and characters that indicate how should a string 
# be formatted:

# \n indicates new line= "H\ne\n\l\nl\no"
print("H\ne\nl\nl\no") # Try this to see the magic
# \t indicates new tab= "H\te\tl\tl\to"
print("H\te\tl\tl\to") # Try this to see the magic

# If you really really really want to use these special characters
# we can use a r in front of the print function in order to print
# also the special characters:

print(r"H \n e \n l \n l \n o") # Everything is printed, right?

# String methods:

# Let's start with some basic examples for converting strings

    # The upper() method: Changes all characters on string to 
    # uppercase:
    
A = "This Has Many Uppercases On the Start of Each Word"
# A has loooots of uppercases, right?, now let's check this next
# method
B = A.upper() # Now what?, try it!
print(B) # Everything is an uppercase right?

    # The lower() method: Changes all characters on string to
    # lowercase:

C = "OMG, I DON'T KNOW HOW TO USE LOWERCASES"
D = C.lower() # Problem solved!
print(D)

    # The replace(x, y) method: Changes characters of a string for others:

E = C.replace('OMG', 'OH MY GOD') # I now want to know what initials 
                                  # stand for! 
    
print(E) # Hmmm... sorry, I'm an atheist, let's change this:
E = E.replace('OH MY GOD', 'OH MY SCIENCE') 
print(E) # Much better.

    # The find(x) method: Indicates where are the specified character 
    # sequence is on a string
    
F = C.find("DON'T") # This will output '7' as 7 is the number of index
                    # on string.
print(F) # Right?
    

```
